```python
from sklearn.cluster import KMeans
x=data[['Channel', 'Region', 'Fresh', 'Milk', 'Grocery', 'Frozen','Detergents_Paper', 'Delicassen']]
```

```python
data.columns
```

```python
#calculate wcss score to calculate numbe of clusters
```

```python
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters = i,n_init=10)
  km.fit(x)
  css = km.inertia_
  wcss.append(css)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('No of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#from elbow method 5 is the optimum no of clusters
km=KMeans(n_clusters=5,n_init=10)
```

```python
km.fit(x)
```

```python
y_km=km.predict(x)
```

```python
y_km
```

```python
pd.Series(y_km).value_counts()
```

```python
data.columns
```

```python
plt.scatter(x['Fresh'],x['Milk'],c = y_km)
plt.xlabel('Fresh')
plt.ylabel('Milk')
plt.show()
```

```python
data['Cluster']=y_km
```

```python
#Agglomerative HC
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings = linkage(x,method = 'centroid')
```

```python
plt.figure(figsize=(10,6))
dendrogram(mergings)
plt.axhline(150,color='red')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean',linkage = 'ward')
y_hc = hc.fit_predict(x)
```

```python
from sklearn.metrics import silhouette_score
silhoutte = silhouette_score(x,y_hc)
print(silhoutte)
```
